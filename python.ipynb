```python
# 4.Check if number is positive, negative or zero
num = float(input("Enter a number: "))

if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Zero")
```

```python
# 5.Factorial of a number

num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact *= i

print("Factorial is:", fact)
```

```python
# 6.Fibonacci series for given range

n=int(input("Enter number of terms:"))

a,b = 0,1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# 8.Purpose of continue statement

# continue is used to skip the current iteration of a loop and move to the next iteration.

# Exa

for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
# 9.Swap two numbers

# With temp variable
a = 5
b = 10

temp = a
a = b
b = temp

print(a, b)

# Without temp variable
a = 5
b = 10

a, b = b, a

print(a, b)
```

```python
# Even or Odd

num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
try:
    num = int(input("Enter a number: "))
    
    if num % 2 == 0:
        print("Even number")
    else:
        print("Odd number")

except ValueError:
    print("Please enter a valid integer!")
```
